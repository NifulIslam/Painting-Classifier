```python
import numpy as np 
import pandas as pd
import os
from fastai import *
!pip install -Uqq fastai fastbook nbdev
from fastbook import *
from fastai.vision.all import *
```

```python
paintings=[
   "Landscape",
    "Portrait",
    "Still Life",
    "Abstract",
    "Impressionist",
    "Cubist",
    "Surrealist",
    "Expressionist",
    "Realist",
    "Pop Art",
]
```

```python
data_path = "data"
if not os.path.exists(data_path): 
  os.mkdir(data_path)

for painting in paintings:

  dest = f"{data_path}/{painting}"
  if not os.path.exists(dest): 
    os.mkdir(dest)

  try:
    painting_urls = search_images_ddg(painting+"painting")
    download_images(dest, urls = painting_urls)
    
  except:
    continue
```
